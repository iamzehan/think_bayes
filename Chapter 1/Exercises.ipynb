# *Exercise 1-1*

* Compute the pobability of Linda being a female banker
  *  In the dataset under the ``'indus10'`` the code for bankers is ``6870``
  *  Under the ``sex`` header ``1==male`` and ``2==female``



````python
"""First we find out the bankers"""
banker=(gss['indus10']==6870)
banker.head()
````

````python
"""Now we find out the females"""
female=(gss['sex']==2)
female.head()
````

````python
"""Now we define functions that accounts the conjunctional and conditional probability for us"""
def prob(A):
  return A.mean()

def conditional(condition,given):
  return prob(condition[given])
````

````python
"""Now we calculate the probability that Linda is a female banker"""
prob(female & banker)
````

* Compute the probability that Linda is a liberal female banker
  * For liberals the range is between ```1``` to ```3``` under ```polviews``` header.
   1.   Extremely Liberal
   2.   Liberal
   3.   Slightly Liberal
  
>First, now we choose the liberals from the dataset. After that we see the probability that she is a liberal female banker.

````python
liberal=(gss['polviews']<=3)
liberal.head()
````

````python
prob(banker & female & liberal)
````
