> # **Conditional Probability is not commutative**


___

We have seen that prob(A & B) == prob(B & A) for conjuntions.

As far as the conditionals go the premises don't culminate in the same manner.

Therein, conditional(A, given= B) != conditional(B, given=A)

```python
conditional(banker,given=female)
```

```python
conditional(female,given=banker)
```

The results imply that majority(77%) of the bankers are females, but the probability of a female being a banker in a cumulative scenario of all female respondents is very low(2%). 

That concludes that conditional probability is not commutative.
___

# **Laws of Probability**


---

There are three theorems:

1. Using conjunction to compute a conditional probability.
2. Using a conditional probability to compute a conjunction.
3. Using conditional(A, B) to compute conditional (B,A)

_Theorem 3 is also know as Bayes' Theorem._

Mathematically:
1. P(A) is the probability of proposition A.
2.P(A & B) is the probability of the conjunction of A and B, that is the probability that both are **True**
3. P(A|B) is the conditional probability of A given that B is **True**. The '|' means given. 


> # **Theorem 1**

```python
female[banker].mean()
```

```python
conditional(female, given=banker)
```

P(A & B) / prob(b)

> # **Theorem 2**

P(A & B) = P(B) P(A|B)

```python
prob(liberal & democrat) == (prob(democrat) * conditional(liberal, given=democrat))
```

> # **Theorem 3: Bayes' Theorem**

We know that conjunction is commutative (means that the probability of vice versa are the same):

  
P(A & B) = P (B & A) (on the contrary P(A|B) $\neq$ P(A|B))


*Applying theorem 2 to the both sides

➡ P(B) P(A|B) = P(A) P(B|A)

➡ P(A|B) = P(A) P(B|A)/P(B) 👈  and That is Bayes' theorem 

```python
conditional(liberal,given=banker) == prob(liberal) * conditional(banker,given=liberal)/prob (banker)
```

> # **The Law of Total Probability**

P(A) = P(B<sub>1</sub> & A) + P(B<sub>2</sub> & A)

In words the total probability of A is the sum of two possibilities: either B<sub>1</sub> and A are True or  B<sub>2</sub> and are True. But this law is only applicable if B<sub>1</sub> and B<sub>2</sub> are:


*   Mutually exclusive (one of them can be True)
*   Collective exhaustive ( one of them must be True)



```python
prob(banker) #P(A)
```

When there are more than two conditions, it is more concise(comprehensive) that we write it as a summation:
